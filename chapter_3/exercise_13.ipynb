```python
"""Chapter 3 | Linear Regression applied exercises."""
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from sklearn.linear_model import LinearRegression
```

```python
# (a) Using the normal() method of your random number generator, create a
# vector, x, containing 100 observations drawn from a N(0,1) distribution.
# This represents a feature, X.

# Create the random number generator
rng = np.random.default_rng(1)

# Generate 100 observations from a N(0, 1) distribution
x = rng.normal(loc=0, scale=1, size=100)

print(x)
```

```python
# (b) Using the normal() method, create a vector, eps, containing 100
# observations drawn from a N(0,0.25) distribution—a normal distribution
# with mean zero and variance 0.25.

# Generate 100 observations from a N(0, 0.25) distribution
eps = rng.normal(loc=0, scale=0.25, size=100)

print(eps)
```

```python
# (c) Using x and eps, generate a vector y according to the model
# Y = −1 + 0.5X + ε. What is the length of the vector y?
# What are the values of β0 and β1 in this linear model?

y = -1 + 0.5 * x + eps
print(y)

# The length of the vector y is 100, and the values of β0 and β1 are -1 and
# 0.5, respectively.
```

```python
# (d) Create a scatterplot displaying the relationship between x and y.
# Comment on what you observe.

plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# (e) Fit a least squares linear model to predict y using x. Comment on the
# model obtained. How do βˆ0 and βˆ1 compare to β0 and β1?
df = pd.DataFrame({'x': x, 'y': y})
results = smf.ols('y ~ x', df).fit()
results.summary()

# The coefficients appear to be similar to the true values of the linear
# equation defined in (c).
```

```python
# (f) Display the least squares line on the scatterplot obtained in (d).
# Draw the population regression line on the plot, in a different color.
# Use the legend() method of the axes to create an appro- priate legend.

model = LinearRegression(fit_intercept = True)
model.fit(x[:, np.newaxis],y)

plt.scatter(df.x, df.y)

xfit = np.linspace(df.x.min(), df.x.max(), 100)
yfit = model.predict(xfit[:, np.newaxis])
fit, = plt.plot(xfit, yfit, color='r')

xpop = np.linspace(df.x.min(), df.x.max(), 100)
ypop = -1 + .5 * xpop
pop, = plt.plot(xpop, ypop, color='g');
plt.legend([fit, pop],['Fit line','Model line'])
```

```python
# (g) Now fit a polynomial regression model that predicts y using x and x2.
# Is there evidence that the quadratic term improves the model fit? Explain
# your answer.

reg = smf.ols('y ~ x + I(x**2)', df).fit()
reg.summary()

# The p-value for the x^2 term is extremely high, which suggests there is no
# evidence to reject the null hypothesis that there is no relationship between
# the x^2 term and the response. Therefore we can claim the added term does
# not improve the model fit.
```

```python
# Repeat (a)–(f) after modifying the data generation process in such a way
# that there is less noise in the data. The model should remain the
# same. You can do this by decreasing the variance of the normal distribution
# used to generate the error term ε in (b). Describe your results.

x = rng.normal(loc=0, scale=1, size=100)
eps = rng.normal(loc=0, scale=0.1, size=100)
y = -1 + 0.5 * x + eps

df = pd.DataFrame({'x': x, 'y': y})
results = smf.ols('y ~ x', df).fit()
results.summary()

# The model fitting to data with less noise appears to have a much higher
# R^2 statistic. This makes sense as the new data set has less variance
# than the original data set.
```

```python
# (i) Repeat (a)–(f) after modifying the data generation process in such a way
# that there is more noise in the data. The model should remain the same.
# You can do this by increasing the variance of the normal distribution used
# to generate the error term ε in (b). Describe your results.

x = rng.normal(loc=0, scale=1, size=100)
eps = rng.normal(loc=0, scale=0.5, size=100)
y = -1 + 0.5 * x + eps

df = pd.DataFrame({'x': x, 'y': y})
results = smf.ols('y ~ x', df).fit()
results.summary()

# The model fitting to data with more noise appears to have a much lower
# R^2 statistic. This makes sense as the new data set has signifcantly
# more variance than the original data set.
```
